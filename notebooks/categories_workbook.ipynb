```python
from bs4 import BeautifulSoup
import os
import re
import requests

BASE_URL = 'https://www.cgtrader.com/3d-models/'

response = requests.get(BASE_URL)

soup = BeautifulSoup(response.content, 'html')

url_model_categories = set()

for link in soup.findAll('a', attrs={'href': re.compile("^https://www.cgtrader.com/3d-models/")}):
    url_model_categories.add(link.get('href'))
```

```python
with open('../data/category_urls.txt', 'w') as f:
    f.write('\n'.join(list(url_model_categories)))
```
